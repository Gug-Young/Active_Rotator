```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
```

```python
X = sym.
```

```python
x = np.linspace(-np.pi,np.pi,10000000)
p = np.linspace(0,1,1000000)[1:-1]
y = norm.pdf(x,0,np.pi/20)# + 1/2*norm.pdf(x,np.pi/2,np.pi/10)
# y = 1/2*norm.pdf(x,-np.pi/2,np.pi/10) + 1/2*norm.pdf(x,np.pi/2,np.pi/10)

plt.plot(x,y)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
```

```python
z1_pdf = np.sum(np.exp(1j*x)*y*(x[1]-x[0]))
z2_pdf = np.sum(np.exp(2j*x)*y*(x[1]-x[0]))
```

```python
np.abs(z1_pdf**2-z2_pdf)
```

```python

plt.plot(x,y)
plt.xlabel(r'$\theta$')
plt.ylabel(r'$f(\theta)$')
plt.title(rf'$|Z_1|={abs(z1_pdf):.02f},|Z_2|={abs(z2_pdf):.02f}=|Z_1^2|={abs(z1_pdf**2):.02f}$')
```

```python
p1 = np.linspace(0,1,10000)[1:-1][::2]
p2 = np.linspace(0,1,10000)[1:-1][1::2]
ppf1 = norm.ppf(p1,-np.pi/2,np.pi/10)
ppf2 = norm.ppf(p2,np.pi/2,np.pi/10)
ppf = np.r_[ppf1,ppf2]

z1_ppf = np.mean(np.exp(1j*ppf))
z2_ppf = np.mean(np.exp(2j*ppf))
```

```python
plt.hist(ppf,bins=100,density=True)
plt.plot(x,y)
```
